```python
col_names = ["Ticker", "COV", "ROI"]
roi_df = pd.DataFrame(columns = col_names)
for ticker in tickers:
    ticker_df = get_df_from_csv(DATA_PATH,ticker)
    sdate,edate = get_valid_dates(ticker_df,S_DATE_STR , E_DATE_STR)
    #print(sdate,edate)
    ticker_fd_df = get_cov_ror(ticker, ticker_df,sdate , edate )
    #print(ticker_fd_df)
    roi_df.loc[len(roi_df.index)] = ticker_fd_df

display(roi_df)
```

```python
#create a prices matrix

base_df = pd.DataFrame()
for ticker in tickers:
    ticker_df = get_df_from_csv(DATA_PATH,ticker)
    sdate,edate = get_valid_dates(ticker_df,S_DATE_STR , E_DATE_STR)
    base_df = merge_df_by_column_name(base_df,ticker_df,ticker,'Adj Close',sdate,edate)

display(base_df)
corr_df = base_df.corr()
display(corr_df)
```

```python
#crecimiento de la inversion sobre un periodo
(base_df / base_df.iloc[0] * 100).plot(figsize=(16, 9))
#retornos
returns = np.log(base_df / base_df.shift(1))
mean_ret = returns.mean() * 252 # 252 average trading days per year
print('retorno promedio en el periodo:')
display(mean_ret)
#display(get_df_from_csv(DATA_PATH, 'INTC'))
```

```python
weights= [100,50,30,10]  #cantidad de acciones po tipo
weights /= np.sum(weights)
print('volatilidad del portfolio: ',np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights))))
```

```python
gsp_df = add_daily_return_to_df(get_df_from_csv(DATA_PATH,'^GSPC'))
risk_free_rate = 0.013
sp_roi = roi_df.loc[roi_df['Ticker']=='^GSPC']['ROI'].values[0]
display(roi_df)
for ticker in tickers:
    base_df = pd.DataFrame()
    sdate,edate = get_valid_dates(gsp_df,S_DATE_STR , E_DATE_STR)
    base_df = merge_df_by_column_name(base_df,gsp_df,'^GSPC','daily_return',sdate,edate)

    ticker_df = add_daily_return_to_df(get_df_from_csv(DATA_PATH,ticker))
    sdate,edate = get_valid_dates(ticker_df,S_DATE_STR , E_DATE_STR)
    base_df = merge_df_by_column_name(base_df,ticker_df,ticker,'daily_return',sdate,edate)
    if ticker != '^GSPC':
        beta = find_beta(base_df)
        expected_ret = risk_free_rate + beta * 0.05
        sharpe =  (expected_ret - risk_free_rate) / (base_df[ticker].std() * 252 ** 0.5)
        roi_ticker = roi_df.loc[roi_df['Ticker']==ticker]['ROI'].values[0]
        
        port_alpha = roi_ticker - risk_free_rate - (beta * (sp_roi - risk_free_rate))
        print('beta ',ticker,' : ',beta,' e_return ',expected_ret, ' sharpe: ',sharpe,'alpha:',port_alpha)
        print(roi_ticker)
```
